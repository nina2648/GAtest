```
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt

import requests

url = "http://www.gutenberg.org/file/74/74-0.txt"
wds = requests.get(url).text

wc = WordCloud(background_color="pink",max_words=800,stopwords=set(STOPWORDS),max_font_size=40,random_state=42)

wc.generate(wds)

plt.imshow(wc)
plt.grid(False)
plt.savefig("wordcloud.png",dpi=600)
plt.show()
```
